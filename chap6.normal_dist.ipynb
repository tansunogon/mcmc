```python
import stan_jupyter as stan
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm 
```

## 6.3 章末問題
## (1)

```python
code = """
data {
    int N;
    real X[N];
}
parameters {
    real mu;
    real<lower=0> sigma;
}
model {
    for (n in 1:N) {
        X[n] ~ normal(mu, sigma);
    }
}
"""

X = [76230, 73550, 80750, 71500, 75420, 74840,
     71580, 76920, 68450, 76990, 64070, 76200]

data = {
    'N': len(X),
    'X': X,
}

posterior = stan.build(code, data=data, random_seed=0)
fit = posterior.sample(num_chains=4, num_samples=100000)
```

```python
fit.to_frame()
```

```python
mu = list(fit.to_frame().mu)
```

```python
mu70000 = list(1 if x > 70000 else 0 for x in mu)
mu75000 = list(1 if x > 75000 else 0 for x in mu)
```

```python
df = pd.DataFrame([mu, mu70000, mu75000],
                  index=["mu", ">70000", ">75000"])
df.T.describe(percentiles=[0.025, 0.975]).T
```

## (2)

```python
code = """
data {
    int N;
    real X[N];
}
parameters {
    real<lower=0> sigma;
}
model {
    for (n in 1:N) {
        X[n] ~ normal(130.0, sigma);
    }
}
"""

X = [128.27, 130.00, 130.68, 128.82, 130.08, 129.24, 129.82,
     131.76, 128.19, 131.01, 128.81, 131.53, 129.36, 130.82,
     130.89, 130.35, 130.18, 131.00, 131.03, 128.84]

data = {
    'N': len(X),
    'X': X,
}

posterior = stan.build(code, data=data, random_seed=0)
fit = posterior.sample(num_chains=4, num_samples=100000)
```

```python
sigma2 = list(sigma**2 for sigma in fit.to_frame().sigma)
```

```python
sigma15 = list(1 if sig2 <= 1.5 else 0 for sig2 in sigma2)
sigma10 = list(1 if sig2 <= 1.0 else 0 for sig2 in sigma2)
```

```python
df = pd.DataFrame([sigma2, sigma15, sigma10],
                  index=["sigma^2", "<=1.5", "<=1.0"])
df.T.describe(percentiles=[0.025, 0.975]).T
```


## (3)

```python
code = """
data {
    int N;
    real X[N];
}
parameters {
    real mu;
    real<lower=0> sigma;
}
model {
    for (n in 1:N) {
        X[n] ~ normal(mu, sigma);
    }
}
"""

X = [75, 82, 77, 86, 98, 91, 85, 84, 82, 79,
     88, 79, 84, 87, 69, 93, 84, 82, 89, 78,
     90, 74, 75, 84, 89, 81, 74, 79, 82, 84]

data = {
    'N': len(X),
    'X': X,
}

posterior = stan.build(code, data=data, random_seed=0)
fit = posterior.sample(num_chains=4, num_samples=100000)
```

```python
mu = list(fit.to_frame().mu)
sigma = list(fit.to_frame().sigma)
```

```python
eta75 = list(mu + 1.282*sigma for mu, sigma in zip(mu, sigma))
biko = list(1 - norm.cdf(boarder, 87, 5) for boarder in eta75)
```

```python
df = pd.DataFrame([eta75, biko],
                  index=["eta", "B子"])
df.T.describe(percentiles=[0.025, 0.975]).T
```

## (4)

```python
X = [30.86, 29.75, 31.55, 32.29, 29.90, 31.71, 31.35, 29.03, 30.37, 31.55,
     29.26, 32.29, 29.90, 30.18, 30.72, 32.28, 30.72, 29.90, 31.55, 31.55]

Y = [31.36, 33.34, 33.16, 31.36, 36.19, 29.80, 31.11, 35.23, 31.36, 31.27,
     31.63, 31.63, 32.00, 31.11, 31.63, 31.36, 31.81, 31.63, 29.21, 33.37]

code = """
data {
    int N;
    real X1[N];
    real X2[N];
}
parameters {
    real mu1;
    real mu2;
    real<lower=0> sigma;
}
model {
    for (n in 1:N) {
        X1[n] ~ normal(mu1, sigma);
        X2[n] ~ normal(mu2, sigma);
    }
}
"""

data = {
    'N': len(X),
    'X1': X,
    'X2': Y,
}

posterior = stan.build(code, data=data, random_seed=0)
fit = posterior.sample(num_chains=4, num_samples=100000)
```

```python
mu1 = list(fit.to_frame().mu1)
mu2 = list(fit.to_frame().mu2)
diff = list(mu2 - mu1 for mu1, mu2 in zip(mu1, mu2))
mu2mu10 = list(1 if mu2 - mu1 > 0 else 0 for mu1, mu2 in zip(mu1, mu2))
mu2mu11 = list(1 if mu2 - mu1 > 1 else 0 for mu1, mu2 in zip(mu1, mu2))
```

```python
df = pd.DataFrame([diff, mu2mu10, mu2mu11],
                  index=["diff", "mu2-mu1>0", "mu2-mu1>1"])
df.T.describe(percentiles=[0.025, 0.975]).T
```

## (5)

```python
X1 = [53.1, 51.5, 45.5, 55.5, 49.6, 50.1, 59.2, 54.7, 53.0, 48.6,
      55.3, 54.6, 51.7, 48.6, 56.4, 58.9, 53.3, 42.4, 51.9, 39.1]
X2 = [51.3, 48.2, 49.6, 59.6, 44.2, 47.6, 54.9, 56.5, 48.4, 50.6,
      53.6, 57.5, 52.0, 46.9, 56.8, 50.0, 53.8, 38.3, 52.6, 41.0]
X = [*zip(X1, X2)]

code = """
data {
    int N;
    vector[2] X[N];
}
parameters {
    vector[2] mu;
    vector<lower=0>[2] sigma;
    real<lower=-1,upper=1> rho;
}
transformed parameters {
    matrix[2,2] Sigma;
    Sigma[1,1] <- pow(sigma[1], 2);
    Sigma[2,2] <- pow(sigma[2], 2);
    Sigma[1,2] <- sigma[1]*sigma[2]*rho;
    Sigma[2,1] <- sigma[1]*sigma[2]*rho;
}
model {
    for (i in 1:N) {
        X[i] ~ multi_normal(mu, Sigma);
    }
}
"""

data = {
    'N': len(X),
    'X': X,
}

posterior = stan.build(code, data=data, random_seed=0)
fit = posterior.sample(num_chains=4, num_samples=100000)
```

```python
fit.to_frame()
```

```python
mu1 = list(fit.to_frame()["mu.1"])
mu2 = list(fit.to_frame()["mu.2"])
diff = list(mu1 - mu2 for mu1, mu2 in zip(mu1, mu2))
mu2mu10 = list(1 if mu1 - mu2 > 0 else 0 for mu1, mu2 in zip(mu1, mu2))
mu2mu12 = list(1 if mu1 - mu2 > 2 else 0 for mu1, mu2 in zip(mu1, mu2))
```

```python
df = pd.DataFrame([diff, mu2mu10, mu2mu12],
                  index=["diff", "mu1-mu2>0", "mu1-mu2>2"])
df.T.describe(percentiles=[0.025, 0.975]).T
```
